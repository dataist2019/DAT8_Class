```python
from sklearn.feature_extraction.text import CountVectorizer
import pandas as pd
import csv
```

```python
url = 'https://raw.githubusercontent.com/jorge3ie/DAT8_Class/master/ier_n.csv'
```

```python
data = pd.read_csv(url)
```

```python
X = data.Title
y = data.label
```

```python
from sklearn.cross_validation import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1)
print X_train.shape
print X_test.shape
```

```python
vect = CountVectorizer(encoding='utf-8', decode_error='ignore', strip_accents='ascii')
```

```python
vect.fit(X_train)
X_train_dtm = vect.transform(X_train)
X_train_dtm
```

```python
X_test_dtm = vect.transform(X_test)
X_test_dtm
```

```python
X_train_tokens = vect.get_feature_names()
```

```python
print X_train_tokens[:50]
```

```python
print X_train_tokens[-50:]
```

```python
X_train_dtm.toarray()
```

```python
import numpy as np
X_train_counts = np.sum(X_train_dtm.toarray(), axis=0)
X_train_counts
```

```python
X_train_counts.shape
```

```python
pd.DataFrame({'token':X_train_tokens, 'count':X_train_counts}).sort('count')
```

```python
data_ham = data[data.label==0]
data_spam = data[data.label==1]
```

```python
data_ham
```

```python
vect.fit(data.Title)
all_tokens = vect.get_feature_names()
```

```python
all_tokens
```

```python
ham_dtm = vect.transform(data_ham.Title)
spam_dtm = vect.transform(data_spam.Title)
```

```python
ham_counts = np.sum(ham_dtm.toarray(), axis=0)
```

```python
spam_counts = np.sum(spam_dtm.toarray(), axis=0)
```

```python
token_counts = pd.DataFrame({'token':all_tokens, 'ham':ham_counts, 'spam':spam_counts})
```

```python
token_counts['ham'] = token_counts.ham + 1
token_counts['spam'] = token_counts.spam + 1
```

```python
token_counts['spam_ratio'] = token_counts.spam / token_counts.ham
token_counts.sort('spam_ratio')
```

```python
from sklearn.naive_bayes import MultinomialNB
nb = MultinomialNB()
nb.fit(X_train_dtm, y_train)
```

```python
y_pred_class = nb.predict(X_test_dtm)
```

```python
from sklearn import metrics
print metrics.accuracy_score(y_test, y_pred_class)
```

```python
print metrics.confusion_matrix(y_test, y_pred_class)
```

```python
y_pred_prob = nb.predict_proba(X_test_dtm)[:, 1]
y_pred_prob
```

```python
print metrics.roc_auc_score(y_test, y_pred_prob)
```

```python
X_test[y_test < y_pred_class]
```

```python
X_test[y_test > y_pred_class]
```

```python
from sklearn.linear_model import LogisticRegression
logreg = LogisticRegression(C=1e9)
logreg.fit(X_train_dtm, y_train)
```

```python
y_pred_class = logreg.predict(X_test_dtm)
y_pred_prob = logreg.predict_proba(X_test_dtm)[:, 1]
```
